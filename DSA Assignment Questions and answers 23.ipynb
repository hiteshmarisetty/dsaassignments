Given aÂ Binary tree, the task is to print theÂ **left view**Â of the Binary Tree. The left view of a Binary Tree is a set of leftmost nodes for every level.

**Examples:**

***Input:***

            4

          / Â  \

        5 Â  Â  2

             / Â  \

            3 Â  Â  1

           / Â \

          6 Â  Â 7

***Output:**Â 4 5 3 6*
***Input:***

                    1

                  / Â  \

                2 Â  Â  Â  3

                 \

                   4

                     \

                        5

                           \

                             6

**Output:**Â 1 2 4 5 6
</aside>

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def leftView(root):
    if root is None:
        return []

    queue = [(root, 0)]
    leftView = []

    while queue:
        node, level = queue.pop(0)

        # Check if it's the first node at the current level
        if level == len(leftView):
            leftView.append(node.value)

        # Add the left and right child nodes to the queue
        if node.left:
            queue.append((node.left, level + 1))
        if node.right:
            queue.append((node.right, level + 1))

    return leftView


# Create the binary tree
root = Node(4)
root.left = Node(5)
root.right = Node(2)
root.right.left = Node(3)
root.right.right = Node(1)
root.right.left.left = Node(6)
root.right.left.right = Node(7)

# Calculate and print the left view
result = leftView(root)
for node in result:
    print(node, end=' ')
```

Given a Binary Tree, print the Right view of it.

TheÂ right view of a Binary TreeÂ is a set of nodes visible when the tree is visited from the Right side.

**Examples:**

**Input:**

         1

      / Â  Â  \

   2 Â  Â   Â  Â 3

/ Â  \ Â  Â  Â  / Â \

4 Â  Â  5 Â  6 Â  Â 7

             \

               8

**Output**: 

Right view of the tree is 1 3 7 8

**Input:**

         1

       /

    8

  /

7

**Output**: 

Right view of the tree is 1 8 7

</aside>

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def rightView(root):
    if root is None:
        return []

    queue = [(root, 0)]
    rightView = []

    while queue:
        node, level = queue.pop(0)

        # Check if it's the first node at the current level
        if level == len(rightView):
            rightView.append(node.value)
        else:
            # Update the rightmost node at the current level
            rightView[level] = node.value

        # Add the left and right child nodes to the queue
        if node.left:
            queue.append((node.left, level + 1))
        if node.right:
            queue.append((node.right, level + 1))

    return rightView


# Create the binary tree
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)
root.right.right.right = Node(8)

# Calculate and print the right view
result = rightView(root)
for node in result:
    print(node, end=' ')
```

Given a Binary Tree, The task is to print theÂ **bottom view**Â from left to right. A nodeÂ **x**Â is there in output if x is the bottommost node at its horizontal distance. The horizontal distance of the left child of a node x is equal to a horizontal distance of x minus 1, and that of a right child is the horizontal distance of x plus 1.

**Examples:**

**Input:**

             20

           / Â  Â  \

        8 Â  Â  Â  Â  22

    / Â  Â  Â \ Â  Â  Â  Â  \

5 Â  Â  Â  Â  3 Â  Â  Â  Â 25

        / Â  Â \

   10 Â  Â  Â  14

**Output:**Â 5, 10, 3, 14, 25.

**Input:**

             20

           / Â  Â  \

        8 Â  Â  Â  Â  22

    / Â  Â  Â \ Â  Â  Â / Â  \

 5 Â  Â  Â  Â  3 Â   4 Â  Â  25

         / Â  Â \

     10 Â  Â  Â  14

**Output:**

5 10 4 14 25.

**Explanation:**

If there are multiple bottom-most nodes for a horizontal distance from the root, then print the later one in the level traversal.

**3 and 4** are both the bottom-most nodes at a horizontal distance of 0, we need to print 4.

</aside>

```python
from collections import deque

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        self.horizontal_distance = 0

def bottomView(root):
    if root is None:
        return []

    # Create a dictionary to store the bottom view nodes
    bottomViewDict = {}

    # Perform a level-order traversal using a queue
    queue = deque()
    queue.append(root)

    while queue:
        node = queue.popleft()
        horizontal_distance = node.horizontal_distance

        # Update the value for the current horizontal distance
        bottomViewDict[horizontal_distance] = node.value

        # Update the horizontal distance for the child nodes
        if node.left:
            node.left.horizontal_distance = horizontal_distance - 1
            queue.append(node.left)
        if node.right:
            node.right.horizontal_distance = horizontal_distance + 1
            queue.append(node.right)

    # Sort the bottom view nodes based on the horizontal distance
    bottomViewNodes = [bottomViewDict[key] for key in sorted(bottomViewDict.keys())]

    return bottomViewNodes


# Create the binary tree
root = Node(20)
root.left = Node(8)
root.right = Node(22)
root.left.left = Node(5)
root.left.right = Node(3)
root.right.right = Node(25)
root.left.right.left = Node(10)
root.left.right.right = Node(14)

# Calculate and print the bottom view
result = bottomView(root)
for node in result:
    print(node, end=' ')
```
